# Plotting

Now that we're familiar with interacting with the data, let's try to visualize it with some plots. We'll use the [matplotlib](https://matplotlib.org/) library to do this.

```python
# imports and data loading
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('realestatedata.csv')
```

```python
# create a simple scatter plot

plt.scatter(df['house_size'], df['price']) # x, y
plt.xlabel('House Size') # x-axis label
plt.ylabel('Price') # y-axis label
plt.title('House Price by Size') # title
plt.show() # show the plot
```

```python
# create a histogram of state frequencies

plt.hist(df['state']) # x
plt.xlabel('State') # x-axis label
plt.ylabel('Frequency') # y-axis label
plt.title('State Frequency') # title
plt.show() # show the plot
```

```python
# the text is overlapping, so let's rotate the x-axis labels

plt.hist(df['state']) # x
plt.xlabel('State') # x-axis label
plt.ylabel('Frequency') # y-axis label
plt.title('State Frequency') # title
plt.xticks(rotation=90) # rotate the x-axis labels
plt.show() # show the plot
```
