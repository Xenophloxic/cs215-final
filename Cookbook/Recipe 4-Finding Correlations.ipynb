# Finding Correlations

Sometimes we want to know if two variables are [correlated](https://en.wikipedia.org/wiki/Correlation_and_dependence). Let's see if there are any correlations between the variables in our dataset.

```python
# import
import pandas as pd
import matplotlib.pyplot as plt

# read data
df = pd.read_csv('realestatedata.csv')
```

```python
# lets look at all correlations

df[['acre_lot', 'price', 'bed', 'bath', 'house_size']].corr()
```

```python
# plot data

# plot price vs size
plt.scatter(df['price'], df['house_size'])
plt.xlabel('Price')
plt.ylabel('Size')
plt.title('Price vs Size')
plt.show()
```

```python
# are there any correlations?

df['price'].corr(df['house_size']) # compute correlation, looks like there is a small correlation
```

```python
# let's plot the correlation line

# plot price vs size, with correlation line using seaborn regplot
import seaborn as sns

# since the data is so large, let's sample it
df_sample = df.sample(1000)

sns.regplot(x='price', y='house_size', data=df_sample)
plt.xlabel('Price')
plt.ylabel('Size')
plt.title('Price vs Size')
plt.show()
```
